# Crear  una función generadora o tambien conocida como variable generador

 Variable /funcion generador: Es una función en programación que permite generar una secuencia de valores de manera iterativa.

En lugar de generar todos los valores de una vez, el generador calcula los valores a medida que se solicitan, lo que puede ser más eficiente en términos de uso de memoria y tiempo de ejecución.



```python
def secciones_datos(datos, tamano_seccion=500):
    for i in range(0, len(datos), tamano_seccion):
        yield ' '.join(datos[i:i+tamano_seccion])
```

# Llamamos a la funcion Generadora  y le pasamos como parametro la variable "a" la cual almacena toda la columna resumen y lo alamcenamos en la variable "generador", se genera un objeto generador .


```python
generador= secciones_datos(a)
print(generador)
```

###Con el metodo "next()" podemos ir solictando los valores al servidor, lo utilizaremos hasta que ya no hayan valores por mostrar.
